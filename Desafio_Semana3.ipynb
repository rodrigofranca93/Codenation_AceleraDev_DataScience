```python
##Importando pacotes
import pandas as pd
import numpy as np
import json
```

```python
#Carregando o dataset
df = pd.read_csv('desafio1.csv')
df.head()
```

```python
df.describe()
```

```python
##Análise Exploratória
df.groupby('estado_residencia')['pontuacao_credito'].describe()
```

```python
##Criando dataframe de resposta
Analise = df.groupby('estado_residencia')['pontuacao_credito'].agg([pd.Series.mode, np.mean, np.median, np.std])
Analise.head()
```

```python
##Renomeando colunas
Analise.rename(columns={'mean':'media','median':'mediana','std':'desvio_padrao','mode':'moda'}, inplace = True)
```

```python
##Gerando arquivo de submissão
Analise.to_json('submission.json',orient = 'index')
```
